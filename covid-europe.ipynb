Features of interest

```python
data_basic = data.loc[:,['Date_reported','Country','WHO_region','New_cases','New_deaths']]
```

```python
data_europe_basic = data_basic[data_basic.Country.isin(['Greece','Italy','Germany','France','Spain'])]
```

```python
data_europe_basic
```

```python
data_europe_basic.set_index('Date_reported')
```

Check for Null items

```python
data_europe_basic.isnull().any()
```

```python
data_grouped = data_europe_basic.groupby('Country')[('New_cases','New_deaths')].sum()
data_grouped
```

```python
death_percentage = (data_grouped.iloc[:,1] / data_grouped.iloc[:,0] ) * 100
```

```python
death_percentage
```

```python
data_grouped['Death_percentage'] = death_percentage
```

```python
data_grouped
```

```python
sns.barplot(x=data_grouped.index, y=data_grouped.Death_percentage)
```

```python
data_grouped.sort_values('New_deaths')
```

```python
date_formated = pd.to_datetime(data_europe_basic.Date_reported)
```

```python
data_europe_basic.Date_reported=date_formated
```

```python
get_year = data_europe_basic.Date_reported.dt.year
get_year
```

```python
group_by_year = data_europe_basic.groupby(['Country',get_year])['New_deaths'].sum()
```

```python
year_frame = group_by_year.to_frame()
```

```python
sns.histplot(data=year_frame, x='New_deaths',hue='Country')
```
